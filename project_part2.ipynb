```python
import geocoder
import project_part1 as p1 #This is part 1 of this project, wherein we scrape the Wiki page for the data
```

```python
df = p1.getDataframe() #Import the data
```

```python
df.head(10)
```

```python
df['Latitude'] = None
df['Longitude'] = None

#Go through each row of the dataframe and get the coordinates of the neighborhood:
for i in range(len(df)):
    lat_lng_coords = None #Initialize the coordinates variable

    #Loop until we get the coordinates:
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(df['PostalCode'][i]))
        lat_lng_coords = g.latlng

    #Set the latitude and longitude values for the current row:
    df['Latitude'].iloc[i] = lat_lng_coords[0]
    df['Longitude'].iloc[i] = lat_lng_coords[1]
```
